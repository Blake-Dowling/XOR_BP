```python
import numpy as np
train_data = [
    [0,0],[0,1],[1,0],[1,1]
]
train_data = np.array(train_data)
train_targets = [0,1,1,0]
train_targets = np.array(train_targets)

W1 = np.array([0.1,0.1])
W2 = np.array([0.1,0.1])

# Sum > 0
def infer1(I):
    O1 = np.dot(W1,I)
    print(O1)
infer1(train_data[1])

```

```python
import matplotlib.pyplot as plt

plt.scatter(train_data[:,0],train_data[:,1],c=train_targets)
plt.plot([0,W1[0]],[W1[1],0])
plt.plot([0,W2[0]],[W2[1],0])
```
